# ğŸ“˜ Veri KÃ¼mesi OluÅŸturma

Bu bÃ¶lÃ¼mde yÃ¶ntemler veri kÃ¼mesi hazÄ±rlanacaktÄ±r.

- Alan ve yÃ¶ntem bilgileri tanÄ±mlanacaktÄ±r.
- Ã–rnek yÃ¶ntem kayÄ±tlarÄ± eklenecektir.
- Veri kÃ¼mesi JSON formatÄ±nda kaydedilecektir.


```python
import json
from datetime import datetime

# YÃ¶ntemler veri kÃ¼mesi listesi
methods = []

def add_method(field, subfield, method_name, description, typical_problems,
               steps=None, success_criteria=None):
    """
    YÃ¶ntem kaydÄ± eklemek iÃ§in yardÄ±mcÄ± fonksiyon.
    """
    method = {
        "id": f"{field.lower().replace(' ', '_')}_{method_name.lower().replace(' ', '_')}",
        "field": field,                      # Bilim alanÄ± / faaliyet alanÄ±
        "subfield": subfield,                # Alt alan (opsiyonel)
        "method_name": method_name,          # YÃ¶ntemin adÄ±
        "description": description,          # KÄ±sa aÃ§Ä±klama
        "typical_problems": typical_problems,# Bu yÃ¶ntemle Ã§Ã¶zÃ¼len problem tÃ¼rleri
        "steps": steps or [],                # AdÄ±m adÄ±m uygulanÄ±ÅŸ (opsiyonel)
        "success_criteria": success_criteria or [],  # BaÅŸarÄ± ne demek? (opsiyonel)
        "created_at": datetime.now().isoformat()
    }
    methods.append(method)

print("YÃ¶ntem veri yapÄ±sÄ± tanÄ±mlandÄ±.")
```
